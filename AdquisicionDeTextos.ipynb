## Probando web scraping con una página

```python
import bs4 as bs # Beautiful Soup
import urllib.request
```

```python
link = "https://es.wikipedia.org/wiki/Alan_Turing"
print(link)
```

```python
# Descargar página web en Python
request = urllib.request.urlopen(link)
source = request.read() 
request.close()
```

```python
pag1 = bs.BeautifulSoup(source,'html.parser') 
```

```python
# title de la etiqueta h1
pag1.find('h1').contents
```

```python
# Se trae el primer parrafo con los links: 
#pag1.find(class_='mw-parser-output').select('.mw-parser-output p')[0].find_all('a')
pag1.find(class_='mw-parser-output').find('p').find_all('a')
```

```python
# Extrayendo las etiquetas del parrafo:

#tags = pag1.find(class_='mw-parser-output').select('.mw-parser-output p')[0].find_all('a')
tags = pag1.find(class_='mw-parser-output').find('p').find_all('a')

for tag in tags:
    print("                                              ")
    print("link:",tag)
    print("contenido:",tag.contents)
    print("titulo:",tag.get_text())
```
